#### Linked List

```python
# Node class
class Node:
    # Function to initialize the node
    def __init__(self, data):
        self.data = data
        self.next = None

    def __repr__(self):
        return self.data
```

```python
# Linked First class
class LinkedList:
    # function to initialize the linkedlist
    def __init__(self):
        self.head = None

    # iterate through all nodes in list
    def __iter__(self):
        node = self.head
        while node is not None:
            yield node
            node = node.next
            
    # display all nodes in list       
    def __repr__(self):
        nodes = []
        for node in self:
            nodes.append(node.data)
        nodes.append("None")
        return "->".join(nodes)

    # push node at the begining of linked list
    def push(self, data):
        # create new node
        newNode = Node(data)
        # new node next to previous head
        newNode.next = self.head
        # new new node as head
        self.head = newNode
       
    # append new node at theend of linked list
    def append(self, data):
        # create new node
        newNode = Node(data)
        
        # if linked list is empty, new node is head
        if self.head is None:
            self.head = newNode
            return
        
        # else add to the end of linked list
        lastNode = self.head
        while(lastNode.next):
            lastNode = lastNode.next
            
        lastNode.next = newNode
        
    # Insert new node after given reference node
    def insertAt(self, prevNode, data):  
        if self.head is None:
            raise Exception("List is empty")
        
        for node in self:
            if node.data == prevNode:
                # create new node
                newNode = Node(data)
                # make next new node as previous node
                newNode.next = node.next
                # make next of prev node as new node
                node.next = newNode
                return
            
        raise Exception("Node with data '%s' not found"% prevNode)
        
    # remove node from list
    def remove(self, data):
        if self.head is None:
            raise Exception("List is Empty")
        if self.head.data == data:
            self.head = self.head.next
            return
        
        prevNode = self.head
        for node in self:
            if node.data == data:
                prevNode.next = node.next
                return
            prevNode = self.head
            
        raise Exception("Node with data '%s' not found"%data)
     
    # count nodes in list
    def count(self):
        count = 0
        for node in self:
            count += 1
        return count
```

```python
# start linked list
linklist = LinkedList()

# start with head
node_first = Node('a')
node_second = Node('b')
node_third = Node('c')

linklist.head = node_first
node_first.next = node_second
node_second.next = node_third

linklist.push('dd')
linklist.append('ww')
linklist.insertAt('c', '2')
print(linklist)

linklist.remove('dd')
print(linklist)
```
